```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('India_GDP.csv')
data
```

```python
data.head()
```

```python
data.notnull()
```

```python
data.tail()
```

```python
data=data.drop(['Unnamed: 0'], axis=1)
```

```python
data
```

```python
 data['Year'].plot(kind='pie')
 plt.title('total number of years [2021-2017]')
 plt.show()
```

```python
#plotting a pie chart
#The pie chart is used to study the proportion of numerical data. 
#It shows the proportion of data as a percentage of a whole.
plt.pie(data['Growth %'], labels = data['Year'])
plt.show()
plt.title('Yearly GDP growth')

#plotted for top 5 countries
df_top5=data.head()
plt.pie(df_top5['Growth %'], autopct='%1.1f%%',labels = df_top5['Year'])
plt.title('Yearly GDP for top 5 countries')
plt.show()
```

```python
# Bar chart is commonly used to compare teh values of variables at a given point of time
data.plot(kind = 'bar',
        y = 'Growth %',
        x= 'Year',
        color = 'purple',width=0.5,figsize=(5,5))
plt.title('BarPlot')
plt.show()
```

```python
#line chart

data.plot(kind = 'line',
        x = 'Year',
        y = 'Growth %',
        color = 'red',marker='o',markerfacecolor = 'green')
plt.show()
```

```python
sns.lineplot(data['Year'],data['Growth %'])
```

```python
#plotting scatterplot using seaborn library
sns.scatterplot(data['Year'],data['Growth %'])
color = 'purple'
plt.title('scatterplot')
plt.show()
```

```python
 data['Growth %'].plot(kind='box')
```

```python
from matplotlib.colorbar import ColorbarBase
colors = sns.color_palette('bright')
#plt.pie(data['Growth %'], labels = data['Year'],colors=colors)
df_top5=data.head()
#We can also add text inside the slices using the autopct parameter.
plt.pie(df_top5['Growth %'], autopct='%1.1f%%',labels = df_top5['Year'])
plt.show()
```

```python
explode = [0.3,0.02,0.02,0.02,0.02]
plt.pie(df_top5['Growth %'], autopct='%1.1f%%',labels = df_top5['Year'],colors=colors,explode=explode)
```

```python
#explode function
#If we want to create space between each slice so that there will be some offset, we can do that using the explode parameter.
#The value of the explode parameter should be an array of the same length as the input data.
explode = [0.3,0.02,0.02,0.02,0.02]
plt.pie(df_top5['Growth %'], autopct='%1.1f%%',labels = df_top5['Year'],colors=colors,explode=explode)
plt.title('BarPlot for Yearly GDP growth')
plt.show()
```

```python
#Bar Chart 
import plotly.express as px
years= list(map(str, range(1960, 2022)))
fig = px.bar(data, x=years,y='Growth %',title= 'Yearly GDP growth')
fig.show()
```

```python
#pie chart using seaborn
#fig = px.pie(Data1['Growth %'],title='Population of European continent')
#fig.show()
fig = px.pie(data, values='Growth %', names=years, title='Yearly GDP Growth')
fig.show()
```

```python
fig = px.line(data, x=years, y="Growth %", title='Yearly GDP Growth')
fig.update_traces(line_color='#456988')
fig.show()
```
